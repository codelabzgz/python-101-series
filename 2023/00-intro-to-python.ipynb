# Listas por compresi贸n
Las **listas por comprensi贸n** son una caracter铆stica de Python y de algunos lenguajes funcionales, que permite especificar una lista en un formato m谩s compacto. En primer lugar, veamos un ejemplo en el que se obtienen los 10 primeros n煤meros impares:

#### Forma "normal":

```python
a = []
for i in range(10):
  a.append(1+i*2)

print(a)
```

#### Forma Python :

```python
b = [2*i+1 for i in range(10)]
print(b)
```

La sintaxis general es `[ f(elem) for elem in list if p(elem) ]`, donde:
- `f` es una funci贸n que se aplica a cada elemento de la lista.
- `elem` es cualquier elemento de la lista `list`.
- `p` es una funci贸n/predicado para determinar si `elem` estar谩 en la lista, es un valor booleano. 

A continuaci贸n veremos otro ejemplo en el que se hace uso de todo lo que ofrecen las listas por compresi贸n:

#### Lista de impares con bucle

```python
c = []
for i in range(20):
  if i%2 != 0:
    c.append(i)

print(c)
```

#### Lista de impares con listas por compresi贸n

```python
d = {i for i in range(20) if i%2 != 0}
print(d)
```

# Clases

Python no es un lenguaje especificamente orientado a objetos, pero permite utilizar un enfoque OOP. En el siguiente ejemplo, se crea una clase `Student` con varios atributos que son inicializados en el **m茅todo** `__init__`. Para declarar m茅todos nuevos se debe escribir una funci贸n en la que el primer par谩metro es `self`, que se corresponde con la instancia que invoca al m茅todo.

```python
class Student:
  fav_lang = "Python"

  def __init__(self, name, age):
    self.name = name
    self.age = age

  def get_name(self):
    return self.name

  def get_age(self):
    return self.age

student1 = Student("Jorge", 19)
print(student1.get_name())
print(student1.get_age())
```

# Ejemplo algoritmo
A continuaci贸n se muestra la belleza del c贸digo Python con el algoritmo de la b煤squeda dicot贸mica.
Este consiste en la busqueda en un vector ORDENADO de un elemento, dado que el vector esta ordenado si el elemento que buscamos es mayor que el actual se encuentra a su derecha, por el contrario si es menor se encuentra a su izquierda.

El algoritmo demuestra la utilidad de los slices principalmente.

```python
def binary_search(arr, target):
    if not arr:
        return -1

    mid = len(arr) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search(arr[mid + 1:], target) + mid + 1
    else:
        return binary_search(arr[:mid], target)

my_list = [1, 3, 5, 7, 9]
target_value = 9
result = binary_search(my_list, target_value)
print(f"Index of {target_value} is {result}")
```
