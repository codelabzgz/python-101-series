# Matplotlib 101

En este notebook puedes probar algunas de las características de la librería Matplotlib.

## Inicializar librerías

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Gráfica de líneas

```python
# Generamos unos datos aleatorios para probar utilizando la librería numpy
data1 = np.sin(0.1 * np.arange(0, 100)) + np.random.uniform(-0.1, 0.1, size=(100,))
data2 = 2 * np.sin(-0.15 * np.arange(0, 100) + 0.6) + np.random.uniform(-0.3, 0.3, size=(100,))

plt.figure()

# mostramos un conjunto de 100 valores
plt.plot(data1, label="data 1")

# podemos mostrar varias líneas en la misma gráfica, con distintos colores
plt.plot(data2, color='red', label="data 2")

# se puede personalizar la gráfica con títulos, leyendas y más
plt.legend()
plt.title("Ejemplo de plot")

plt.show()
```

## Puntos

```python
# más datos aleatorios: 1000 números entre 0 y 100
x1size = 1000
x2size = 200
x1 = np.random.uniform(0, 100, (x1size,))
x2 = np.random.uniform(0, 100, (x2size,))

y1 = np.sin(0.1 * x1) + np.random.uniform(-0.2, 0.2, size=(x1size,))
y2 = 2 * np.cos(0.1 * x2) + np.random.uniform(-0.1, 0.1, size=(x2size,))

plt.figure()

# mostramos un conjunto de 100 valores
plt.scatter(x1, y1, label="data 1")

# podemos mostrar varias líneas en la misma gráfica, con distintos colores
plt.scatter(x2, y2, label="data 2")

# se puede personalizar la gráfica con títulos, leyendas y más
plt.legend()
plt.title("Ejemplo de plot")

plt.show()
```
